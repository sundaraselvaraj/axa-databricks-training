### Component Reference

| Component | Description |
|---|---|
| **Azure Resource Group** | Logical container that holds all related Azure resources for the Databricks deployment. |
| **Databricks Workspace** | The top-level environment where you create notebooks, manage clusters, schedule jobs, and explore data. |
| **ADLS Gen2** | Azure Data Lake Storage Gen2 is the recommended storage layer for data lakes, providing hierarchical namespace and fine-grained access control. |
| **Azure Key Vault** | Centralized secrets management; used to store connection strings, passwords, and API keys securely. |
| **Control Plane** | Managed by Databricks in their own Azure tenant. It hosts the web UI, job scheduler, REST APIs, and cluster management. |
| **Data Plane** | Runs inside **your** Azure subscription. This is where actual compute (VMs), storage, and data processing happen. Your data never leaves your subscription. |

## Section 3 &mdash; Create Your First Notebook

Notebooks in Databricks are collaborative documents that combine code, visualizations, and narrative text.

### How to create a notebook

1. Navigate to **Workspace > Users > your_username > labs**.
2. Click the **Create** button (or right-click) and select **Notebook**.
3. Give it a name, select **Python** as the default language, and attach it to a running cluster.

## Section 4 &mdash; Magic Commands

One of the most powerful features of Databricks notebooks is the ability to mix multiple languages in a single notebook using **magic commands**. A magic command starts with `%` at the very beginning of a cell.

```python
displayHTML("""
<table style="font-family: 'Segoe UI', Arial, sans-serif; border-collapse: collapse; width: 100%; max-width: 800px; margin: 10px auto; border: 1px solid #E8E5E0;">
  <tr>
    <th style="background: #1B3A4B; color: #FFFFFF; padding: 12px 16px; text-align: left; font-size: 0.95em;">Magic Command</th>
    <th style="background: #1B3A4B; color: #FFFFFF; padding: 12px 16px; text-align: left; font-size: 0.95em;">Language / Action</th>
    <th style="background: #1B3A4B; color: #FFFFFF; padding: 12px 16px; text-align: left; font-size: 0.95em;">Description</th>
  </tr>
  <tr style="background: #FFFFFF;"><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;"><code>%python</code></td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">Python</td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">Default in Python notebooks; run Python code</td></tr>
  <tr style="background: #F9F7F4;"><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;"><code>%sql</code></td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">SQL</td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">Run Spark SQL queries directly</td></tr>
  <tr style="background: #FFFFFF;"><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;"><code>%scala</code></td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">Scala</td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">Execute Scala code on the JVM <em>(not on Serverless)</em></td></tr>
  <tr style="background: #F9F7F4;"><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;"><code>%r</code></td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">R</td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">Run R statistical computations</td></tr>
    <tr style="background: #F9F7F4;"><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;"><code>%fs</code></td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">DBFS Utilities</td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">Shorthand for <code>dbutils.fs</code> commands <em>(not on Serverless — use dbutils.fs directly)</em></td></tr>
  <tr style="background: #FFFFFF;"><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;"><code>%md</code></td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">Markdown</td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">Render Markdown text inside a code cell</td></tr>
  <tr style="background: #F9F7F4;"><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;"><code>%pip</code></td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">Pip</td><td style="padding: 10px 16px; border-bottom: 1px solid #E8E5E0; color: #1B1F24;">Install Python libraries on the cluster</td></tr>
  <tr style="background: #FFFFFF;"><td style="padding: 10px 16px; color: #1B1F24;"><code>%run</code></td><td style="padding: 10px 16px; color: #1B1F24;">Run notebook</td><td style="padding: 10px 16px; color: #1B1F24;">Execute another notebook and import its variables</td></tr>
</table>
""")
```

## Section 6 &mdash; Creating and Querying Tables

Databricks lets you create managed tables using SQL. In this section, you will create a sample database, populate it with data, and run queries.

### 6.1 &mdash; Create a Database

```python
%sql
CREATE DATABASE IF NOT EXISTS lab01_db;
SHOW DATABASES
```

### 6.2 &mdash; Create a Table with Sample Data

```python
%sql
USE lab01_db;

CREATE OR REPLACE TABLE employees (
  id INT,
  name STRING,
  department STRING,
  salary DOUBLE,
  hire_date DATE
);

INSERT INTO employees VALUES
  (1, 'Alice Martin', 'Engineering', 95000, '2021-03-15'),
  (2, 'Bob Dupont', 'Marketing', 82000, '2020-07-01'),
  (3, 'Claire Laurent', 'Engineering', 98000, '2019-11-20'),
  (4, 'David Chen', 'Data Science', 105000, '2022-01-10'),
  (5, 'Eva Schmidt', 'Marketing', 78000, '2023-04-05');
```

### 6.3 &mdash; Query the Table

```python
%sql
SELECT * FROM lab01_db.employees ORDER BY salary DESC
```

### 6.4 &mdash; Aggregations with GROUP BY

```python
%sql
SELECT
  department,
  COUNT(*)        AS employee_count,
  ROUND(AVG(salary), 2) AS avg_salary,
  MIN(salary)     AS min_salary,
  MAX(salary)     AS max_salary
FROM lab01_db.employees
GROUP BY department
ORDER BY avg_salary DESC
```
